```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
os.path.exists('student_grades - Sheet1.csv')
```

```python
grades = pd.read_csv('student_grades - Sheet1.csv') 
grades
```

```python
grades.dtypes
```

```python
grades.student_id = grades.student_id.astype(object)
grades.exam2 = grades.exam2.astype(float)
grades.final_grade = grades.final_grade.astype(float)
grades.dtypes
```

```python
grades.isna().sum()
```

```python
grades.exam1 = grades.exam1.fillna(round(grades.exam1.mean(), 1))
grades.exam3 = grades.exam3.fillna(round(grades.exam3.mean(), 1))
grades.isna().sum()
```

```python
sns.pairplot(grades, corner=True)
plt.show()
```
